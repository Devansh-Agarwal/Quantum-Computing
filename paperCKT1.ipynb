```python
import qiskit as qk

# Creating Qubits
qr=qk.QuantumRegister(12)
cr=qk.ClassicalRegister(12)
tr=qk.QuantumRegister(2)
tc=qk.ClassicalRegister(1,'c')
qc= qk.QuantumCircuit(qr,tr,cr,tc)


qc.x(qr[2])##### location of the prize

qc.h(qr[6])
qc.h(qr[7])
qc.ch(qr[6],qr[7])
#qc.x(qr[0:2])
#qc.reset(qr[6])
#qc.x(qr[0])



qc.x(qr[7:10])
qc.mct(qr[8:10],tr[0],None)#control, target, none
qc.mct([tr[0],qr[7]],tr[1],None)
qc.cz(tr[1],qr[6])
qc.reset(tr[0:2])
#qc.x(qr[7:10])


#qc.x(qr[7:10])
qc.mct(qr[8:10],tr[0],None)#control, target, none
qc.mct([tr[0],qr[7]],tr[1],None)
qc.ch(tr[1],qr[6])
qc.reset(tr[0:2])
qc.x(qr[8])



qc.mct(qr[8:10],tr[0],None)#control, target, none
qc.mct([tr[0],qr[7]],tr[1],None)
qc.ch(tr[1],qr[6])
qc.reset(tr[0:2])
qc.x([qr[7],qr[9]])


qc.x([qr[6],qr[8]])
qc.mct(qr[8:10],tr[0],None)#control, target, none
qc.mct([tr[0],qr[6]],tr[1],None)
qc.ch(tr[1],qr[7])
qc.reset(tr[0:2])
qc.x([qr[6],qr[8]])

# 2nd part ended
qc.x([qr[1],qr[6]])
qc.mct([qr[1],qr[6]],tr[0],None)
qc.ch(tr[0],qr[7])
qc.reset(tr[0:2])
qc.x([qr[1],qr[6]])


qc.x(qr[7])
qc.mct([qr[1],qr[7]],tr[0],None)
qc.ch(tr[0],qr[6])
qc.reset(tr[0])
qc.x(qr[7])


qc.x([qr[0],qr[6]])
qc.mct([qr[0],qr[6]],tr[0],None)
qc.cz(tr[0],qr[7])
qc.ch(tr[0],qr[7])
qc.reset(tr[0])
qc.x([qr[0],qr[6]])


qc.x([qr[2],qr[7]])
qc.mct([qr[0],qr[2]],tr[0],None)#control, target, none
qc.mct([tr[0],qr[7]],tr[1],None)
qc.ch(tr[1],qr[6])
qc.reset(tr[0:2])
qc.x([qr[2],qr[7]])



qc.x([qr[1],qr[6]])
qc.mct([qr[0],qr[1]],tr[0],None)#control, target, none
qc.mct([tr[0],qr[6]],tr[1],None)
qc.ch(tr[1],qr[7])
qc.reset(tr[0:2])
qc.x([qr[1],qr[6]])




qc.cx(qr[0],qr[3])
qc.cx(qr[1],qr[4])
qc.cx(qr[2],qr[5])
qc.cx(qr[6],qr[10])
qc.cx(qr[7],qr[11])


# last part
qc.x(qr[11])
qc.ch(qr[11],qr[10])
qc.x(qr[10])
qc.mct([qr[11],qr[10]],tr[0],None)

qc.measure(tr[0],tc[0])
qc.measure([qr[0],qr[1]],[cr[0],cr[1]]).c_if(tc,1)
#if(tc[0]==1):
#	print('lolo1')
#	qc.measure([qr[0],qr[1]],[cr[0],cr[1]])
qc.reset(tr[0])


qc.x(qr[10])
qc.ch(qr[11],qr[10])
qc.x(qr[11])
qc.x(qr[10])
qc.mct([qr[10]],tr[0],None)

qc.measure(tr[0],tc[0])
qc.measure([qr[2],qr[3]],[cr[2],cr[3]]).c_if(tc,1)
#if(tc[1]==1):
#	print('lolo2')
#	qc.measure([qr[2],qr[3]],[cr[2],cr[3]])	

qc.reset(tr[0])
qc.measure(tr[0],tc[0])


qc.x(qr[10])
qc.cx(qr[11],qr[10])
qc.mct([qr[10]],tr[0],None)

qc.measure(tr[0],tc[0])
qc.measure([qr[4],qr[5]],[cr[4],cr[5]]).c_if(tc,1)
#if(tc[2]==1):
#	print('lolo3')
#	qc.measure([qr[4],qr[5]],[cr[4],cr[5]])	
qc.reset(tr[0])

qc.measure(tr[0],tc[0])


simulator = qk.BasicAer.get_backend('qasm_simulator')

# Simulating the circuit using the simulator to get the result
job = qk.execute(qc, simulator)
result = job.result()

# Getting the aggregated binary outcomes of the circuit.
counts = result.get_counts(qc)
print (counts)
```
